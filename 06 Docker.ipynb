```python
'''
Micro-services

If a large application is broken down into smaller services, each of those services or small processes can be termed 
micro-services and they communicate with each other over a network.

The microservices approach is the opposite of the monolithic approach which can be difficult to scale. If one particular 
feature has some issues or crashes, all other features will experience the same.

Another example is when the demand for a particular feature is seriously increasing, we are forced to increase the resources 
such as the hardware not only for this particular feature but for the entire application generating additional costs that 
are not necessary. 
```

```python
'''
Recall:

Dcoker:
    Docker is a tool used to automate the deployment of an application as a lightweight container so that the application can 
    work efficiently in different environments.
    
    Docker is a containerization platform which packages your application and all its dependencies together in the form of 
    containers so as ensure that our application works seamlessly in any environment, be it development, test or production. 
    
Docker Container:
    Docker Containers are sort of encapsulated environments in which you run applications.
    Docker containers include the application and all of its dependencies. and ship it all out as one package. 
    
    It shares the kernel with other containers, running as isolated processes in user space on the host operating system.
    Docker Container run on any computer, on any infrastructure, and in any cloud. Docker containers are basically runtime 
    instances of Docker images.
    
    Docker will take care of shiping this containers to all the possible platform in standad way.
    
    Important states of Docker container are:
      - Running
      - Paused
      - Restarting
      - Exited

Docker image?
    A Docker image is a template that allows you to instantiate running containers. It is represented as a list of instructions 
    (known as layers) in a filesystem.
    
    Base Image : It Says that what is Basic OS you want to have.
    
Docker Hub
    Docker Hub is the place where base images will be available

Dockerfile?
    A Dockerfile is just a blueprint(text document)that contains the instructions to build a Docker image. Currently, more than 
    a million Dockerfiles are on GitHub.

Docker Engine?
    Docker daemon or Docker engine represents the server. The docker daemon and the clients should be run on the same or 
    remote host, which can communicate through command-line client binary and full RESTful API.
    
Docker Daemon: 
    - A persistent background process that manages Docker images, containers, networks, and storage volumes.
    - The Docker daemon constantly listens for Docker API requests and processes them.   
    
Docker Registry?
  - A registry is a storage and content delivery system, holding named Docker images
  - Users interact with a registry by using docker push and pull commands.

Docker build
    The process of building a Docker image from a Dockerfile is known as a Docker build.
    Users can pick up images from Docker Hub and use them to create customized images and containers

Docker layer
    Each layer in a Docker context represents an instruction included in a Docker image's Dockerfile. 
    The layers can also be referred to as "build steps".

Docker build cache?
    Every time you build a Docker image, each build step is cached. Reuse cached layers that do not change in the image rebuild 
    process to improve the build time.

Docker Swarm
    - Docker Swarm is an orchestration management tool that runs on Docker applications.
    - It helps end-users in creating and deploying a cluster of Docker nodes.
    
Docker Compose
    Docker Compose is a YAML file which contains details about the services, networks, and volumes for setting up the Docker 
    application.
    So, you can use Docker Compose to create separate containers, host them and get them to communicate with each other. 
    Each container will expose a port for communicating with other containers.
    
Docker Namespace
    Namespaces provide a layer of isolation.Docker provides various namespaces in order to stay portable and not affect the 
    underlying host system.
    
Docker Machine
    Docker machine is a tool that lets you install Docker Engine on virtual hosts. These hosts can now be managed using the 
    docker-machine commands. Docker machine also lets you provision Docker Swarm Clusters.

Docker Network
    - Docker networking allows you to attach a container to as many networks as you like
    - Default is Bridge; Host; None is for isolation 
```

```python
'''
Docker Swarm:
    - Docker Swarm is an orchestration management tool that runs on Docker applications.
    - It helps end-users in creating and deploying a cluster of Docker nodes.
    - It allows the user to manage multiple containers deployed across multiple host machines.
    
What Dcoker Swarm contains
    - In a docker swarm, there are typically several worker nodes and at least one manager node that is responsible for 
      handling the worker nodes' resources efficiently and ensuring that the cluster operates efficiently.

Types of Docker Swarm mode services?
    - Replicated and Global
    
Docker Swarm Nodes
    - Docker Swarm Manager Node
        - The primary function of manager nodes is to assign tasks to worker nodes in the swarm.
        - Docker recommends a maximum of seven manager nodes for a swarm
        
    - Docker Swarm Leader Node
        -When a cluster is established, the Raft consensus algorithm is used to assign one of them as the "leader node". 
        
    - Docker Swarm Worker Node
        - In a docker swarm with numerous hosts, each worker node functions by receiving and executing the tasks that are 
          allocated to it by manager nodes.
```
