```python
'''
I/O Modules:
    - IO-Link enables bidirectional communication from the control level down to the field device (sensor, actuator). 
    - I/O modules are building blocks of industrial control systems used in end equipment such as PLC 
    - Analog input modules capture and measure signals such as temperature, pressure, flow, level, vibration and motion. These 
      parameters are transmitted from equipment such as sensors and other field devices on the factory floor in the form of 
      current or voltage. Current loops typically range from 4-20mA while voltage inputs can be single ended or differential 
      ranging from 0-5V, 0-10V, +/-5V and +/-10V.
      
AC 800M controllers(ABB DCS Device):PLC Device
    The AC 800M controller is a family of rail-mounted modules, consisting of CPUs, communication modules, power supply modules 
    and various accessories. Several CPU modules are available that vary in terms of processing power, memory size, SIL-rating, 
    and redundancy support.
    

Control Builder
    - Efficient control system engineering and configuration
    - Control Builder can be used to configure AC 800M control and safety applications.
    - Control Builder is a powerful tool for creating control solutions and reusable control libraries for the AC 800M 
      controller.
      
800xA History Server RTDB (OPC Server)
    - The RTDB is a proprietary ABB Database.
    - It is a Proprietary Data Base, developed by ABB. However, additionally the data can be accessed via SQL commands, either 
      directly via an SQL Client or by utilizing and ODBC Client.
    - Also, there are other methods to access the data. The latest version of 800xA History contains an OPC-UA Server for DA 
      and HDA data access and and OPC Server for A/E data access. 
```
